```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
```

```python
# Tentukan versi tokenizer yang ingin Anda gunakan
tokenizer_version = "gpt2"

# Muat tokenizer
tokenizer = GPT2Tokenizer.from_pretrained(tokenizer_version)
```

```python
# Muat model
model = GPT2LMHeadModel.from_pretrained("gpt2")
```

```python
# Fungsi untuk menghasilkan teks
def generate_text(prompt, max_length=50):
    input_ids = tokenizer.encode(prompt, return_tensors="pt")
    output = model.generate(input_ids, max_length=max_length, num_return_sequences=1)
    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
    return generated_text
```

```python
# Contoh penggunaan aplikasi
prompt = "tell me about love"
generated_text = generate_text(prompt, max_length=1000)
print("Generated Text:")
print(generated_text)
```
