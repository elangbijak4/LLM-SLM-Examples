```python
from transformers import BertTokenizer, BertModel
import torch
```

```python
# Inisialisasi tokenizer dan model BERT
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
```

```python
# Contoh teks
text = "The quick brown fox jumps over the lazy dog."
```

```python
# Tokenisasi teks
inputs = tokenizer(text, return_tensors="pt")
```

```python
# Mendapatkan output dari model BERT
with torch.no_grad():
    outputs = model(**inputs)
```

```python
# Mengambil representasi embedding dari token [CLS] (token awal) dari layer terakhir
last_hidden_states = outputs.last_hidden_state
cls_embedding = last_hidden_states[:, 0, :]
```

```python
# Hasilkan vektor embedding dari tensor PyTorch menjadi array numpy
embedding = cls_embedding.numpy()
```

```python
print("Embedding shape:", embedding.shape)
```
